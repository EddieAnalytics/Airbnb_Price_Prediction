# Define X and y for the prediction model


*   take a sample size from the dataset for both X and y
*   split the sample size into training data and test data
*   define the parameter for gradientboostRegressor
*   fine tune the Regressor

The version shown below is the best outcome. For this dataset, there are many variables that affect the model's performance. 

If we slice the dataset by categories such as industries, we may be able to make individual models for each industry to better predict the salary of a new job. Right now, it is more generalized. 


```python
#define the x variable
#Now that I am done with exploratory analysis and data cleaning. 
#I am dropping the salary column and jobId column for X here as they are not needed
X = df_encoded.drop(['jobId','salary'],axis = 1).copy()
X.head()
```

```python
#name the target variable
#I am saving the salary column as y 
y = df_encoded['salary'].copy()
y.head()
```

```python
#I tried to train a a set of data without the jobtype janitor. But the results are still very similar
X_noJ = df.drop(df[df['jobType']=='JANITOR'].index)
y_noJ = X_noJ['salary'].copy()
X_nJ = X_noJ.drop(['jobId','salary'],axis = 1).copy()

X6 = X.sample(n = 60000, random_state = 0)
y6 = y.sample(n = 60000, random_state = 0)

X6_train, X6_test, y6_train, y6_test = train_test_split(X6, y6, test_size = 0.3, random_state=0)
```

```python
#Below is the number of samples that I am choosing for the training model. 
#We have a large dataset. It would be too slow to run through every single row. 
#setting the random states the same here so that we have the same rows.
X1 = X.sample(n = 60000, random_state = 0)
y1 = y.sample(n = 60000, random_state = 0)
```

```python
#split the dataset
#Using the train_test_split package to split the data to avoid overfitting
X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size = 0.3, random_state=0)

X
```

```python
#create a validation dataset to check on overfitting
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=0)
```

```python
#Use the ensemble learning's gradient boosting regressor to predict the salary
#There are a lot of parameters to specify in this model
#n_estimators:the number of sequential trees to be molded 
model_g = ensemble.GradientBoostingRegressor(
    n_estimators = 900, #n_estimators:the number of sequential trees to be molded 
    learning_rate = 0.05, #shrinks the contribution of each tree by this rate
    max_depth = 5, #how deep the trees can be
    min_samples_leaf = 10, #min number of samples required to split an internal node
    max_features = 0.6, 
    loss='huber', #refers to the loss function to be minimized in each split
    random_state = 0 #ensures the results are consistent
)

model_g.fit(X_train, y_train)

g_mae = MAE(y_train, model_g.predict(X_train))
print("Training Set MAE:%4f" % g_mae)
g_mse = MSE(y_test, model_g.predict(X_test))
print("MSE:%4f" % g_mse)
print("RMSE Score: ", np.sqrt(g_mse))
print("R2 Score: {:.4f}".format(r2_score(y_test, model_g.predict(X_test))))
```

```python
#model_g = ensemble.GradientBoostingRegressor()

#this parameter is just to explore which one would generate the best results
#Once the results are generated by fitting the model,
#The best parameter tells us how we should adjust our numbers in the dictionary
param_grid = {
    'n_estimators':[900,1000,1100],
    'learning_rate':[0.05,0.06, 0.07],
    'max_depth':[5,6,7],
    'min_samples_leaf':[7,8],
    'max_features':[0.2,0.3,0.4],
    'loss':['huber']
}

gb_grid = GridSearchCV(model_g, param_grid, n_jobs =4)

gb_grid.fit(X_train, y_train)

print("The best hyper parameter setting is {}".format(gb_grid.best_params_))
```

```python
#check the model's score, also the accuracy
print(model_g.score(X_test, y_test))
```

```python
#Name the importance of the features
#then sort the importance
importances = model_g.feature_importances_ 
indices = np.argsort(importances)[::-1] 
print("Top 10 Important Features\n" )
#for each feature importance in the top 15 of them,
for f in range(15): 
    #print out each of the feature's importance
    print("%d. %s (%f)" % (f + 1, X.columns[indices[f]], importances[indices[f]])) 
    #Plot the feature importances of the forest 
    indices=indices[:15] 
#plot a figure that has all of the 15 features using barplot
#rotate the xticks for readability
plt.figure()
plt.title("Top 15 Feature importances") 
plt.bar(range(15), importances[indices], color="orange", align="center") 
plt.xticks(range(15), X.columns[indices], fontsize=10, rotation=90) 
plt.xlim([-1, 15]) 
plt.show() 
#Mean Feature Importance 
print("Mean Feature Importance %.2f"% np.mean(importances)) 
```

# XGBoost
I did not use xgboost but still kept it for reference because I think it is a great model with more computing power. It just takes too long to run on a personal computer, with such a large dataset, and multiple parameters to consider. 

I did not run the part of the code that fine tuness the xgboost regressor because it would take too long to generate results. 



```python
#define the training, validation, and test values using xgboost
#xgb.DMatrix converts the data into the format that xgboost recognizes
#Here we have the training set, validation set, and the test set. 
dtrain = xgb.DMatrix(X_train, label=y_train)
dval = xgb.DMatrix(X_val, label=y_val)
dtest = xgb.DMatrix(X_test, label=y_test)
```

```python
#Setting the parameters seperately so that I can modify later to tune the RMSE score. 
#the learning rate shrinks the feature weights to make the boosting process more conservative
#max_depth is the max number of trees. The more max_depth we have, the more likely the model is going to overfit. 
#Colsample_bytree is the fraction of randomly selected features that will be used to train each train
#random_state makes sure that our results are consistent
#There are other parameters such as colsampe_bylevel
#I did not add them as they significantly slow down the processing time
params = {'learning_rate':0.001,
          "colsample_bytree":0.6,
          'max_depth':4,
          'random_state':0
          
}

model = xgb.train(params, dtrain, evals=[(dval, 'eval')], num_boost_round = 20000, early_stopping_rounds=10)
```

```python
#model.get_score(importance_type='gain', max_num_features=10)
#the plot below shows the importance and the impact it has on the salary.
#As we can see that milesFromMetropolis is the most important variable, followed by yearsExperience
#I only showed the most important 15 variables for illustraion
xgb.plot_importance(model, max_num_features = 15)
plt.show()
```

```python
#Here I am putting the y_test value and the model's prediction of the test set into arrays
#Then I will later use them to calculate the RMSE score
y_true = np.array(y_test)
y_pred = np.array(model.predict(dtest))
```

```python
#Here I am checking what the R2 score
#The R2 score is the 1-(sum of square of residuals (errors)/sum of square of total) 
print("R2 Score: {:.4f}".format(r2_score(y_true, y_pred)))
```

```python
#Print the measuring metrics for this model
print("XGBoost score on training set: ", np.sqrt(mean_squared_error(y_test, y_pred)))
print("XGBoost score on training set: ", mean_squared_error(y_test, y_pred))
```
